```python
import re
```

```python
with open('homicides.txt',mode='rU') as f:
    data=[row for row in f]
    
```

```python
ages=[]
for row in data:
    match=re.search(r'\d+ years? old',row)
    if match:
        ages.append(match.group())
    else:
        ages.append('0')
        

    
```

```python
# split the string on spaces, only keep the first element, and convert to int
ages=[int(age.split()[0]) for age in ages]
ages
```

```python
# calculate average age
sum(ages)/len(ages)
```

```python
causes=[]
with open('homicides.txt',mode='rU') as f:
    data=[row for row in f]
for cause in data:
    match=re.search(r'Cause:\s\w+<',cause)
    if match:
        str1=match.group().replace('<','').replace("Cause:",'').replace(' ','')
        causes.append(str1)
    else:
        causes.append('ok')
        
len(data)
```

```python
from collections import Counter
Counter(causes)
```
